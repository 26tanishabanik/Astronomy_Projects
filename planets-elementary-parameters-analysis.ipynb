```python
data = pd.read_csv('GSOC_Data_DataCube.txt', names = ['optical_depth', 'optical_depth_uncertainity', \
                                                        'max_expansion_velocity', 'max_expansion_velocity_uncertainity',\
                                                       'emergent_flux', 'total_mass', 'nickle_mass',\
                                                       'explosion_energy', 'mass_distribution_flag',\
                                                       'nickel_radial_distribution_flag'], sep=" ")
```

```python
data.head()
```

```python
params = {
          "text.color" : "w",
          "ytick.color" : "w",
          "xtick.color" : "w",
          "axes.labelcolor" : "w",
          "axes.edgecolor" : "w"
          }
plt.rcParams.update(params)
```

```python
data.drop(['optical_depth_uncertainity', 'max_expansion_velocity_uncertainity'], axis = 1, inplace = True)
```

```python
from sklearn.preprocessing import OrdinalEncoder
ord_enc = OrdinalEncoder()
```

```python
data["nickel_radial_distribution_flag_code"] = ord_enc.fit_transform(data[["nickel_radial_distribution_flag"]])
data["mass_distribution_flag_code"] = ord_enc.fit_transform(data[["mass_distribution_flag"]])
data.drop(['mass_distribution_flag', 'nickel_radial_distribution_flag'], axis = 1, inplace = True)
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1,11):
    k_means = KMeans(n_clusters=i,init='k-means++', random_state=42, algorithm = 'elkan')
    k_means.fit(data.iloc[:,3:8].values)
    wcss.append(k_means.inertia_)
#plot elbow curve
plt.plot(np.arange(1,11),wcss)
plt.xlabel('Clusters')
plt.ylabel('SSE')
plt.show()
```

```python
X = data.iloc[:,3:8].values
```

```python
k_means_optimum = KMeans(n_clusters = 2, init = 'random',  random_state=42, algorithm = 'elkan')
y = k_means_optimum.fit_predict(X)
print(y)
```

```python
data['cluster'] = y 
```

```python
data1 = data[data.cluster==0]
data2 = data[data.cluster==1]
# data3 = data[data.cluster==2]
# data4 = data[data.cluster==3]
# data5 = data[data.cluster==4]
# data6 = data[data.cluster==5]
```

```python
plt.figure(figsize=(15,8))
kplot = plt.axes(projection='3d')
xline = np.linspace(0, 15, 1000)
yline = np.linspace(0, 15, 1000)
zline = np.linspace(0, 15, 1000)
# kplot.plot3D(xline, yline, zline, 'black')

kplot.scatter3D(data1.optical_depth, data1.max_expansion_velocity, data1.emergent_flux, c='red', label = 'Cluster 1')
kplot.scatter3D(data2.optical_depth,data2.max_expansion_velocity,data2.emergent_flux,c ='green', label = 'Cluster 2')
# kplot.scatter3D(data3.optical_depth, data3.max_expansion_velocity, data3.emergent_flux, c='yellow', label = 'Cluster 3')
# kplot.scatter3D(data4.optical_depth, data4.max_expansion_velocity, data4.emergent_flux, c='blue', label = 'Cluster 4')
# kplot.scatter3D(data5.optical_depth, data5.max_expansion_velocity, data5.emergent_flux, c='pink', label = 'Cluster 5')
# kplot.scatter3D(data6.optical_depth, data6.max_expansion_velocity, data6.emergent_flux, c='blue', label = 'Cluster 6')
plt.scatter(k_means_optimum.cluster_centers_[:,0], k_means_optimum.cluster_centers_[:,1], color = 'indigo', s = 200)
plt.legend()
plt.title("Kmeans")
plt.show()
```

```python
from sklearn.metrics import silhouette_score
score = silhouette_score(X,y)
print(score)
```
